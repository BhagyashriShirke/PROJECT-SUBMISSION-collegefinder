# multiplication table 

```python
num = int(input("Multiplication using value? : "))    #table start from which number
                                                      
while (num<=12):
    i=1                     
    while (i<=10):
        product=num*i
        print(num, "*" ,i, "=",product)
        i=i+1
    print("\n")
    num=num+1
```
